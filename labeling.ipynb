```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
nltk.download("vader_lexicon")
import pandas as pd
data = pd.read_csv("/content/data0.31.csv")
data = data.dropna()
print(data.head())
```

```python
sentiments = SentimentIntensityAnalyzer()
data["Yes"] = [sentiments.polarity_scores(i)["neg"] for i in data["Teks"]]
data["No"] = [sentiments.polarity_scores(i)["pos"] for i in data["Teks"]]
data["Compound"] = [sentiments.polarity_scores(i)["compound"] for i in data["Teks"]]

data.head()
```

```python
data
```

```python
score = data["Compound"].values
sentiment = []
for i in score:
  if i >= 0.00 :
    sentiment.append('No')
  else:
      sentiment.append('Yes')
data["label"] = sentiment
data
```

```python
data.to_excel('data0.32.xlsx', index = False)
```
